```python
# prompt: Utilizar modelo cargado

from tensorflow.keras.models import load_model
import cv2, os, pprint
import matplotlib.pyplot as plt
import numpy as np

model = load_model('model10_best_model.h5')
# You can now use the 'model' object for predictions or further training.
# For example, to see the model summary:
# agregar pesos
model.load_weights('model10_best.weights.h5')
model.summary()
```

```python
# Esta variable contiene un mapeo de número de clase a personaje.
# Se utilizan sólo los 18 personajes del dataset que tienen más imágenes.
MAP_CHARACTERS = {
    0: 'abraham_grampa_simpson', 1: 'apu_nahasapeemapetilon', 2: 'bart_simpson',
    3: 'charles_montgomery_burns', 4: 'chief_wiggum', 5: 'comic_book_guy', 6: 'edna_krabappel',
    7: 'homer_simpson', 8: 'kent_brockman', 9: 'krusty_the_clown', 10: 'lisa_simpson',
    11: 'marge_simpson', 12: 'milhouse_van_houten', 13: 'moe_szyslak',
    14: 'ned_flanders', 15: 'nelson_muntz', 16: 'principal_skinner', 17: 'sideshow_bob'
}

# Se estandarizan todas las imágenes a tamaño 64x64
IMG_SIZE = 64
```

```python
# funcion para pasar una imagen a un formato adecuado para la predicción y luego predecir
def predict_image(model, image_path):
    # Cargar la imagen
    img = cv2.imread(image_path)
    img = cv2.resize(img, (64, 64))
    img = img / 255.0  # Normalizar la imagen
    img = np.expand_dims(img, axis=0)  # Añadir una dimensión para el batch size

    # Predecir la clase
    pred = model.predict(img)
    class_index = np.argmax(pred[0])
    return MAP_CHARACTERS[class_index]
```

```python
# mostrar imagen y prediccion
def show_image_and_prediction(image_path, model):
    img = cv2.imread(image_path)
    img = cv2.resize(img, (64, 64))
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    pred_class = predict_image(model, image_path)
    plt.title(f"Predicción: {pred_class}")
    plt.show()
```

```python
image_paths = [
    "homelo_chino.jpg",
    "homero_colocolino.jpg",
    "homero.jpg",
    "homero2.jpg"
]

# Mostrar imagenes y predicciones
for image_path in image_paths:
    print(f"Predicción para {image_path}:")
    show_image_and_prediction(image_path, model)
```
